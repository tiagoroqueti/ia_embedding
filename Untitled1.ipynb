```python
DOCUMENT1 = {
    "Titulo": "Os jogadores",
    "Conteudo": "As equipes são compostas por 11 jogadores, incluindo 1 goleiro. Entre os jogadores, um é o capitão, sendo responsável pela conduta da sua equipe."
}
DOCUMENT2 = {
    "Titulo": "A determinação do resultado de um jogo",
    "Conteudo": "A equipe que tiver marcado mais gols vence a partida de futebol. Se as duas equipes marcarem o mesmo número de gols, o jogo fica empatado, podendo ser desempatado através de pênaltis ou dois tempos iguais de prorrogação (quinze minutos cada tempo, mais possível tempo de acréscimo)."
}
DOCUMENT3 = {
    "Titulo": "O impedimento",
    "Conteudo": "O impedimento acontece quando qualquer parte de um jogador (com exceção dos braços e mãos) está mais perto da linha do gol adversário do que a bola e o penúltimo adversário no momento em que a bola é passada para ele."
}
DOCUMENT4 = {
    "Titulo": "O penalti",
    "Conteudo": "Um penalti é dado quando um jogador cometer uma infração punível com um tiro livre direto dentro da própria área penal. Com exceção do batedor e do goleiro, os jogadores devem ficar a, pelo menos, 9,15 metros de distância da marca penal (fora da área). O pênalti só pode ser cobrado após o apito do árbitro."
}

documents = [DOCUMENT1,DOCUMENT2,DOCUMENT3,DOCUMENT4]
```

```python
df = pd.DataFrame(documents)
df.columns = ["Titulo", "Conteudo"]
df
```

```python
model = "models/embedding-001"
```

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                                 content=text,
                                 title=title,
                                 task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: embed_fn(row["Titulo"],row["Conteudo"]), axis=1)
df
```

```python
def gerar_e_buscar_consulta(consulta, base, model):
  embedding_da_consulta = genai.embed_content(model=model,
                                 content=consulta,
                                 task_type="RETRIEVAL_QUERY")["embedding"]
  produtos_escalares = np.dot(np.stack(df["Embeddings"]), embedding_da_consulta)

  # Define um limiar para a similaridade
  limiar_similaridade = 0.7

  # Verifica se o maior produto escalar está acima do limiar
  if np.max(produtos_escalares) > limiar_similaridade:
    indice = np.argmax(produtos_escalares)
    return df.iloc[indice]["Conteudo"]
  else:
    return "Não sei a resposta."

```

```python
consulta = "Como é um pênalti?"

trecho = gerar_e_buscar_consulta(consulta, df, model)

print(trecho)
```

```python
generation_config = {
    "temperature": 0,
    "candidate_count": 1,
}
```

```python
prompt = f"Reescreva esse texto de uma forma mais descontraída, sem adicionar informações que não façam parte do texto: {trecho}"

model_2 = genai.GenerativeModel("gemini-1.0-pro",
                                generation_config=generation_config)
response = model_2.generate_content(prompt)
print(response.text)
```
